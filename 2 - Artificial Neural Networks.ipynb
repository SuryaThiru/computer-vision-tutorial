# Introduction to Artificial Neural Networks

**Artificial neural networks (ANN)** or connectionist systems are computing systems vaguely inspired by the biological neural networks that constitute animal 
brains. They are capable of learning to perform specific tasks. Nodes are connected through edges which have a weight that adjusts as learning proceeds.

Signal(typically real numbers) flows through the network get transformed across layers and produces output.

**Goal of neural networks**: Find weights that minimizes the error between expected output and the generated output

## The Keras framework

Keras is a high-level neural networks API. Allows fast experimentation

```
from sklearn.datasets import load_iris

dataset = load_iris()
print(dataset['DESCR'])
```

```
X = dataset['data']
Y = dataset['target']
```

```
from keras import Sequential
from keras.layers import Dense, Activation
from keras.activations import hard_sigmoid

model = Sequential()

model.add(Dense(units=10, input_shape=(4,), activation='relu'))
model.add(Dense(units=3, activation='softmax'))

model.summary()
```

```
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```
model.fit(X, Y, epochs=200)
```
